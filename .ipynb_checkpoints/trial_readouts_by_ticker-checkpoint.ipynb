```python
import sqlite3
import pandas as pd
import os

# ── CONFIG ──────────────────────────────────────────────────────────────────
DB_PATH  = 'C:\\Users\\chris\\Documents\\personal_projects\\biotech.db'
TICKER   = 'RVMD'   # <── change this to any ticker
MONTHS_AHEAD = 24  # how far out to look
# ────────────────────────────────────────────────────────────────────────────

os.chdir(os.path.dirname(DB_PATH))
conn = sqlite3.connect(DB_PATH)

query = """
    WITH latest_trial AS (
        SELECT *,
               ROW_NUMBER() OVER (PARTITION BY nct_id ORDER BY snapshot_date DESC) AS rn
        FROM trials
    ),
    latest_pub AS (
        SELECT nct_id, title AS latest_publication, pub_date,
               ROW_NUMBER() OVER (PARTITION BY nct_id ORDER BY pub_date DESC) AS rn
        FROM publications
    ),
    ticker_map AS (
        -- resolve the ticker back to the company name stored in trials
        SELECT DISTINCT ticker, normalized_name
        FROM company_map
        WHERE UPPER(ticker) = UPPER(:ticker)
    )
    SELECT
        t.nct_id,
        t.company,
        t.sponsor,
        t.phase,
        t.status,
        t.conditions,
        t.interventions,
        t.primary_completion_date,
        t.enrollment,
        t.primary_outcomes,
        p.latest_publication,
        p.pub_date
    FROM latest_trial t
    LEFT JOIN latest_pub p
        ON t.nct_id = p.nct_id AND p.rn = 1
    WHERE t.rn = 1
      AND t.phase IN ('PHASE2', 'PHASE3', 'PHASE4')
      AND t.status IN ('ACTIVE_NOT_RECRUITING', 'RECRUITING', 'ENROLLING_BY_INVITATION')
      AND t.primary_completion_date BETWEEN date('now') AND date('now', '+' || :months || ' months')
      AND (
          UPPER(t.company) = UPPER(:ticker)
          OR t.company IN (SELECT normalized_name FROM ticker_map)
      )
    ORDER BY t.primary_completion_date
"""

df = pd.read_sql_query(query, conn, params={'ticker': TICKER, 'months': MONTHS_AHEAD})
conn.close()

pd.set_option('display.max_colwidth', 80)
pd.set_option('display.max_rows', 200)

print(f"Ticker: {TICKER.upper()}  |  Trials found: {len(df)}  |  Horizon: {MONTHS_AHEAD} months")
print()
```

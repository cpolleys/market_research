```python
import sqlite3
import pandas as pd
import os

# ── CONFIG ──────────────────────────────────────────────────────────────────
DB_PATH  = 'C:\\Users\\chris\\Documents\\personal_projects\\biotech.db'
TICKER   = 'LLY'   # <── change this to any ticker
MONTHS_AHEAD = 24  # how far out to look
# ────────────────────────────────────────────────────────────────────────────

os.chdir(os.path.dirname(DB_PATH))
conn = sqlite3.connect(DB_PATH)

query = """
    WITH latest_trial AS (
        SELECT *,
               ROW_NUMBER() OVER (PARTITION BY nct_id ORDER BY snapshot_date DESC) AS rn
        FROM trials
    ),
    latest_pub AS (
        SELECT nct_id, title AS latest_publication, pub_date,
               ROW_NUMBER() OVER (PARTITION BY nct_id ORDER BY pub_date DESC) AS rn
        FROM publications
    ),
    ticker_map AS (
        -- resolve the ticker back to the company name stored in trials
        SELECT DISTINCT ticker, normalized_name
        FROM company_map
        WHERE UPPER(ticker) = UPPER(:ticker)
    )
    SELECT
        t.nct_id,
        t.company,
        t.sponsor,
        t.phase,
        t.status,
        t.conditions,
        t.interventions,
        t.primary_completion_date,
        t.enrollment,
        t.primary_outcomes,
        p.latest_publication,
        p.pub_date
    FROM latest_trial t
    LEFT JOIN latest_pub p
        ON t.nct_id = p.nct_id AND p.rn = 1
    WHERE t.rn = 1
      AND t.phase IN ('PHASE2', 'PHASE3', 'PHASE4')
      AND t.status IN ('ACTIVE_NOT_RECRUITING', 'RECRUITING', 'ENROLLING_BY_INVITATION')
      AND t.primary_completion_date BETWEEN date('now') AND date('now', '+' || :months || ' months')
      AND (
          UPPER(t.company) = UPPER(:ticker)
          OR t.company IN (SELECT normalized_name FROM ticker_map)
      )
    ORDER BY t.primary_completion_date
"""

df = pd.read_sql_query(query, conn, params={'ticker': TICKER, 'months': MONTHS_AHEAD})
conn.close()

pd.set_option('display.max_colwidth', 80)
pd.set_option('display.max_rows', 200)

print(f"Ticker: {TICKER.upper()}  |  Trials found: {len(df)}  |  Horizon: {MONTHS_AHEAD} months")
print()
```

```python
# ── DIAGNOSTIC: understand exactly what's in your DB ──────────────────────
import sys, os, sqlite3
import pandas as pd

sys.path.insert(0, os.path.dirname(DB_PATH))
from mappings import clean_name

diag_conn = sqlite3.connect(DB_PATH)

print('=== company_map rows for this ticker ===')
cm2 = pd.read_sql_query(
    'SELECT raw_name, normalized_name, ticker, confidence FROM company_map '
    'WHERE UPPER(ticker) = UPPER(:t) ORDER BY confidence DESC',
    diag_conn, params={'t': TICKER}
)
print(cm2.to_string(index=False) if not cm2.empty else '  (none found)')

# Derive cleaned names the same way the pipeline does
_cleaned = [clean_name(r) for r in cm2['raw_name']] if not cm2.empty else []
print(f'\nCleaned name(s) pipeline would store: {_cleaned}')

print()
print('=== company values in trials (name-based search) ===')
companies = pd.read_sql_query(
    'SELECT DISTINCT company, COUNT(*) AS trial_count FROM trials GROUP BY company ORDER BY company',
    diag_conn
)
search_term = _cleaned[0].split()[0] if _cleaned else TICKER.lower()
mask = companies['company'].str.lower().str.contains(search_term, na=False)
print(companies[mask].to_string(index=False) if mask.any() else f'  (no match for "{search_term}")')

print()
print('=== all trials for matched company (any status/phase) ===')
if _cleaned:
    ph = ','.join(['?' for _ in _cleaned])
    all_trials = pd.read_sql_query(
        f'SELECT nct_id, phase, status, primary_completion_date, snapshot_date '
        f'FROM trials WHERE LOWER(company) IN ({ph}) '
        f'ORDER BY snapshot_date DESC, primary_completion_date',
        diag_conn, params=[n.lower() for n in _cleaned]
    )
    print(all_trials.to_string(index=False) if not all_trials.empty else '  (no trials found)')
else:
    print('  (could not resolve company name — ticker not in company_map)')

diag_conn.close()
```

```python
# ── DIAGNOSTIC: see how the company is stored in your DB ────────────────────
# Run this cell if the query above returned nothing.

diag_conn = sqlite3.connect(DB_PATH)

print("=== company_map rows for this ticker ===")
cm = pd.read_sql_query(
    "SELECT raw_name, normalized_name, ticker, confidence FROM company_map "
    "WHERE UPPER(ticker) = UPPER(:t) ORDER BY confidence DESC",
    diag_conn, params={'t': TICKER}
)
print(cm.to_string(index=False) if not cm.empty else "  (none found)")

print()
print("=== distinct company values in trials table that fuzzy-match ===")
companies = pd.read_sql_query(
    "SELECT DISTINCT company, COUNT(*) AS trial_count "
    "FROM trials GROUP BY company ORDER BY company",
    diag_conn
)
needle = TICKER.lower()
mask   = companies['company'].str.lower().str.contains(needle, na=False)
print(companies[mask].to_string(index=False) if mask.any() else "  (no match by name)")

diag_conn.close()
```
